#### Product names also include storage/RAM information, which should be preserved because they represent different product variants

```python
def create_base_product(product_name):
    """
    Creates a standardized product name by:
    - Keeping the phone model
    - Keeping RAM information
    - Keeping Storage information
    - Removing all colour/marketing names
    """

    if pd.isna(product_name):
        return product_name

    product_name = product_name.strip()

    # Model name = everything before first bracket
    model = product_name.split("(")[0].strip()

    # Extract RAM
    ram_match = re.search(
        r'(\d+\s*GB\s*RAM|\d+\s*GB)',
        product_name,
        re.IGNORECASE
    )

    # Extract Storage
    storage_match = re.search(
        r'(\d+\s*GB\s*(?:Storage)?)',
        product_name,
        re.IGNORECASE
    )

    # Find all GB values
    gb_values = re.findall(r'\d+\s*GB', product_name, re.IGNORECASE)

    ram = None
    storage = None

    if "RAM" in product_name.upper():

        ram = re.search(
            r'(\d+\s*GB)\s*RAM',
            product_name,
            re.IGNORECASE
        )

        ram = ram.group(1) if ram else None

        storage = re.search(
            r'(\d+\s*GB)\s*(Storage)?',
            product_name,
            re.IGNORECASE
        )

        if storage:
            storage = storage.group(1)

    else:

        if len(gb_values) >= 2:
            ram = gb_values[0]
            storage = gb_values[1]

        elif len(gb_values) == 1:
            storage = gb_values[0]

    specs = []

    if ram:
        specs.append(f"{ram} RAM")

    if storage:
        specs.append(storage)

    if specs:
        return f"{model} ({', '.join(specs)})"

    return model
```

```python
df["base_product"] = df["product_name"].apply(create_base_product)

df[["product_name", "base_product"]].head(30)
```
